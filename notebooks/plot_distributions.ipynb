```python
import numpy as np
import matplotlib.pyplot as plt
```

Sentence length transformation

```python
sentence_length = 15
alpha = -1.2
sp = 1
wealth = 10
arrest_aversion = 1

transformed_sentence_length = sp / (sentence_length ** (1 / alpha))
print(f"Transformed sentence length: {transformed_sentence_length}")

# print the expected punishment pain to compare the two methods
expected_punishment_pain = wealth + arrest_aversion * transformed_sentence_length
print(f"Transformed expected punishment pain: {expected_punishment_pain}")

old_expected_punishment_pain = wealth + arrest_aversion * sentence_length
print(f"Straight expected punishment pain: {old_expected_punishment_pain}")
```

```python
# plot the sentence length
x = np.arange(1, 100)
y = sp / (x ** (1 / alpha))

plt.plot(x, y)
plt.title("Transformed sentence length vs sentence length")
plt.xlabel("Sentence length")
plt.ylabel("Transformed sentence length")
# fix the y axis to 100
plt.ylim(0, 100)
plt.xlim(0, 100)
# plot a line trough the origin
plt.plot([0, 100], [0, 100], color='red')
plt.show()

# plot a table of sentence length and transformed sentence length
print("Sentence length | Transformed sentence length")
for i in range(1, 30):
    print(f"{i} | {sp / (i ** (1 / alpha))}")
```
